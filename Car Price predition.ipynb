```python
FinalData=pd.DataFrame

x=RawData.drop(['Price','New_Price'],axis=1)
y=RawData.drop(x.columns,axis=1)

# print(x['Transmission'].unique())
# x['Fuel_Type']=x['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2,'LPG':3,'Electric':4})
# x['Owner_Type']=x['Owner_Type'].map({'First':0,'Second':1,'Third':2,'Fourth & Above':3})
# x['Transmission']=x['Transmission'].map({'Manual':0,'Automatic':1})
x['Mileage']=pd.to_numeric(x['Mileage'].str.split(" ",expand=True)[0],errors='coerce')
x['Engine']=pd.to_numeric(x['Engine'].str.split(" ",expand=True)[0],errors='coerce')
x['Power']=pd.to_numeric(x['Power'].str.split(" ",expand=True)[0],errors='coerce')
for i in x.columns:
    print(i," : ", x[i].dtype)
x['Mileage'].fillna(x['Mileage'].value_counts().idxmax(),inplace=True)
x['Engine'].fillna(x['Engine'].value_counts().idxmax(),inplace=True)
x['Power'].fillna(x['Power'].value_counts().idxmax(),inplace=True)
x['Seats'].fillna(x['Seats'].value_counts().idxmax(),inplace=True)
# print(x['Mileage'].str.split(" ",expand=True))
print("\n\n\n")
for i in x.columns:
    print(i," : ", x[i].dtype)


# """For mapping Name"""
x['ManufacturerName']= x['Name'].str.split(" ",expand=True)[0]
# ManufacturNameAll=x['Name'].str.split(" ",expand=True)[0] 
# print(ManufacturName)
# DicttForName={}
# i=0
# for j in ManufacturName:
#     DicttForName[j]=i
#     i+=1
# i=0    
# # print(dictt)
# x['Name']=x['Name'].str.split(' ',expand=True)[0]
# x['Name']=ManufacturNameAll.map(DicttForName)
x.drop('Location',axis=1,inplace=True)
# """For mapping City"""
# # DictForLocation={}
# for j in x['Location'].str.split(" ",expand=True)[0].unique():
#     DictForLocation[j]=i
#     i+=1
# print(DictForLocation)
# x['Location']=x['Location'].map(DictForLocation)
# x['Location']=pd.to_numeric(x['Location'].str.split(" ",expand=True)[0].unique())

# display(x['Mileage'].value_counts().idxmax())
# display(x)
# print(x.isna().sum())



# for i in x.columns:
#     print(x[i].dtype)

dummies1=pd.get_dummies(x['ManufacturerName'])
dummies2=pd.get_dummies(x['Fuel_Type'])
dummies3=pd.get_dummies(x['Owner_Type'])
dummies4=pd.get_dummies(x['Transmission'])
print(dummies1.columns,dummies2.columns,dummies3.columns,dummies4.columns)
print(dummies1.shape)
print(dummies2.shape)
print(dummies3.shape)
print(dummies4.shape)
# dummiesForFuel=pd.get_dummies(x['Fuel_Type'])
```

## Year


```python
from datetime import datetime
d=datetime.now().year
x['Year'].fillna(x['Year'].value_counts().idxmax(),inplace=True)
x['Age']=d-x['Year']
x.drop('Year',axis=1,inplace=True)
print(x['Age'].isna().sum())

# display(x)
pd.DataFrame(x.columns).to_csv('XColums.csv',index=False)
```
